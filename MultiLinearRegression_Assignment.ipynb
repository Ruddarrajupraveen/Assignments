```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv("C:/Users/praveen/Multiple_Linear_Regression_Advertising_Data.csv")
```

```python
x=data.drop('Sales', axis=1)
y=data.Sales
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y)
```

```python
MultiRegression=LinearRegression()
```

```python
MultiRegression.fit(x_train,y_train)
```

```python
predictions=MultiRegression.predict(x_test)
```

```python
predictions[1]
```

```python
y_test[1]
```

```python
r2_score(y_test,predictions)
```

```python
plt.xlabel("'Tv','Radio','Newspaper'")
plt.ylabel('Sales')
plt.plot(x,y)
plt.title("MultipleLinearRegression")
```
